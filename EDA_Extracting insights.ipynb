```python
from pyspark.sql.functions import col

df_retails_silver = df_transaction.join(df_customers,"customer_id").join(df_products,"product_id").join(df_store, "store_id").withColumn("total_amount", col("quantity") * col("price"))
display(df_retails_silver)
```

```python
df_retails_silver.write.mode("overwrite").saveAsTable("mohamedDB.sales.retails_silver")
```

```python
from pyspark.sql.functions import col, sum , count,month
```

```python
# Group by country and sum the total_amount
df_country_sales = df_retails_silver.groupBy("country").agg(sum("total_amount").alias("Total_Sales"))


# Display the result
display(df_country_sales)

# Save the aggregated data as a gold table
df_country_sales.write.mode("overwrite").saveAsTable("mohamedDB.sales.df_country_sales_gold")
```

```python
df_country_customer = df_retails_silver.groupBy("country").agg(count("customer_id").alias("Count_Of_Customrs"))
display(df_country_customer)
df_country_customer.write.mode("overwrite").saveAsTable("mohamedDB.sales.df_country_customer_gold")
```

```python
# Group by store_name and sum the total_amount
df_store_sales = df_retails_silver.groupBy("store_name").agg(sum("total_amount").alias("Total_Sales"))

# Display the result
display(df_store_sales)
df_store_sales.write.mode("overwrite").saveAsTable("mohamedDB.sales.df_store_sales_gold")
```

```python
# Group by category and sum the total_amount
df_category_sales = df_retails_silver.groupBy("category").agg(sum("total_amount").alias("Total_Sales_by_category"))

# Display the result
display(df_category_sales)
df_category_sales.write.mode("overwrite").saveAsTable("depi.sales.category_sales_gold")
```

```python
from pyspark.sql.functions import month, sum

df_monthly_sales = df_retails_silver.groupBy(
    month("transaction_date").alias("Month")
).agg(
    sum("total_amount").alias("Total_Sales_per_month")
)

display(df_monthly_sales)

df_monthly_sales.write.mode("overwrite").saveAsTable("mohamedDB.sales.monthly_sales_gold")
```
